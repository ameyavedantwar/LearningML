```python
#Find the Independant and Dependant Features
#Industry practice: X = Independant Features Y = Dependant Features
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
#Standardise the Dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
#Why only transform over here?
#To avoid data leakage = myu will be same if we use fit_transform in
# X_train, X_test which will lower the variance of the test dataset
#means less variety of questions in exam for model
```

```python
#Model Training
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
#Print the coefficiencts and intercepts
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```

```python
#Prediction
reg_predict = regressor.predict(X_test)
```

```python
print(reg_predict)
```

```python
#Assumptions of Linear Regression
```

```python
plt.scatter(y_test,reg_predict)
plt.xlabel("Test Trusth value")
plt.ylabel("Test predicted value")
```

```python
#Calculate Residual
residual = y_test - reg_predict
```

```python
residual
```

```python
sns.displot(residual,kind="kde")
```

```python
#Scatter plot with prediction and residual
#uniform destribution
plt.scatter(reg_predict,residual)
plt.xlabel("Predicted")
plt.ylabel("Error")
```

```python
#The above graph shows it is not inclined towards anywhere which means model is correct 
#homosedacity
```

```python
#Performance Matrix
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_predict))
print(mean_absolute_error(y_test, reg_predict))
print(np.sqrt(mean_squared_error(y_test,reg_predict)))
```

```python
#R-square and Adjusted R-square
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_predict)
print(score)
```

```python
#Adjusted R square
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
