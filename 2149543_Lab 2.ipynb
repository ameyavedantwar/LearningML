```python
import numpy as np
df_data = np.load("D:/1_Swansea/1_Uni/Big data and machine learning/Labs/Lab 2/Iris_data.npy")
df_label = np.load("D:/1_Swansea/1_Uni/Big data and machine learning/Labs/Lab 2/Iris_labels.npy")
```

```python
print(df_data)
print(df_label)
```

```python
'''
The Iris dataset consist of 4 features sepal length, 
sepal width, petal length, petal width of 50 samples 
of 3 species of Iris which are Iris Setosa, Iris Verginica, 
and Iris Versicolor.

These features gives a brief description of differentiating
factors of different Iris types and gives an overall intuition
to the model to seprate it into different clusters.

'''
```

```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
```

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 1)
    kmeans.fit(df_data,df_label)
    wcss.append(kmeans.inertia_)
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(df_data,df_label)
gt = kmeans.predict(df_data)
print(y_kmeans)
print(gt)
```

```python
#Visualising the clusters
plt.scatter(df_data[y_kmeans == 0, 0], df_data[y_kmeans == 0, 1], s = 20, c = 'teal', label = 'Iris-setosa')
plt.scatter(df_data[y_kmeans == 1, 0], df_data[y_kmeans == 1, 1], s = 20, c = 'purple', label = 'Iris-versicolour')
plt.scatter(df_data[y_kmeans == 2, 0], df_data[y_kmeans == 2, 1], s = 20, c = 'yellow', label = 'Iris-virginica')
plt.gca().update(dict(title='Clustering Lab 2', xlabel='x', ylabel='y'))
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 20, c = 'red', label = 'Centroids')
plt.legend()
```

```python
from sklearn.mixture import GaussianMixture 


gmm = GaussianMixture(n_components = 3)
gmm.fit(df_data)
y = gmm.predict(df_data)
print(y)
```

```python
plt.scatter(df_data[y == 0, 0], df_data[y == 0, 1], s = 20, c = 'teal', label = 'Iris-setosa')
plt.scatter(df_data[y == 1, 0], df_data[y == 1, 1], s = 20, c = 'purple', label = 'Iris-versicolour')
plt.scatter(df_data[y == 2, 0], df_data[y == 2, 1], s = 20, c = 'yellow', label = 'Iris-virginica')
plt.gca().update(dict(title='Clustering Gaussian Lab 2', xlabel='x', ylabel='y'))
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 40, c = 'red', label = 'Centroids', marker = 'X')
plt.legend()
```
